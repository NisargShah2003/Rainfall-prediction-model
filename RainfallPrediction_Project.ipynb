```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("weatherAUS.csv")
data
```

```python
df = pd.DataFrame(np.random.randint(0,10,(1,23)))

for i in range(2006,2017):
    df1 = data[((str(i) + '-10-31') < data['Date']) & (data['Date'] <= (str(i+1) +'-04-30'))]
    df = pd.concat([df,df1],axis=0,join="inner")
        
print(df)
```

```python
x = list(df.index)
x
```

```python
data1 = []
for i in x:
    data1.append(data.iloc[i])
df1 = pd.DataFrame(data1)
df1.drop(0,inplace=True)
df1
```

```python
data.info()
```

```python
df1['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
df1['RainTomorrow'].replace({'No': 0, 'Yes': 1},inplace = True)
```

```python
from sklearn.utils import resample
import matplotlib.pyplot as plt

no = df1[df1.RainTomorrow == 0]
yes = df1[df1.RainTomorrow == 1]
yes_oversampled = resample(yes, replace=True, n_samples=len(no))
oversampled = pd.concat([no, yes_oversampled])

fig = plt.figure(figsize = (8,5))
oversampled.RainTomorrow.value_counts(normalize = True).plot(kind='bar', color= ['skyblue','navy'], alpha = 0.9, rot=0)
plt.title('RainTomorrow Indicator No(0) and Yes(1) after Oversampling (Balanced Dataset)')
plt.show()

print("Size of the data after oversampling:",oversampled.shape)
```

```python
# Detecting outliers with IQR

df1 = oversampled.copy(deep=True)
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
Q1
```

```python
# Removing outliers from the dataset

df1 = df1[~((df1.lt((Q1 - 1.5 * IQR))) |(df1.gt((Q3 + 1.5 * IQR)))).any(axis=1)]
df1.shape
```

```python
# Get list of categorical variables

s = (df1.dtypes == "object")
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
# Missing values in categorical variables

for i in object_cols:
    print(i, df1[i].isnull().sum())
```

```python
for i in object_cols:
    df1[i].fillna(df1[i].mode()[0], inplace=True)
```

```python
# Get list of neumeric variables

t = (df1.dtypes == "float64")
num_cols = list(t[t].index)

print("Neumeric variables:")
print(num_cols)
```

```python
# Missing values in numeric variables

for i in num_cols:
    print(i, df1[i].isnull().sum())
```

```python
# Filling missing values with median of the column in value

for i in num_cols:
    df1[i].fillna(df1[i].median(), inplace=True)
    
df1.info()
```

```python
from sklearn.preprocessing import LabelEncoder
# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
for i in object_cols:
    df1[i] = label_encoder.fit_transform(df1[i])
    
df1.info()
```

```python
df1
```

```python
# Standardizing data

from sklearn import preprocessing
r_scaler = preprocessing.MinMaxScaler()
r_scaler.fit(df1)
df1 = pd.DataFrame(r_scaler.transform(df1), index=df1.index, columns=df1.columns)
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier as rf

x = df1.drop('RainTomorrow', axis=1)
y = df1['RainTomorrow']
selector = SelectFromModel(rf(n_estimators=100, random_state=0))
selector.fit(x, y)
support = selector.get_support()
features = x.loc[:,support].columns.tolist()
print(features)
print(rf(n_estimators=100, random_state=0).fit(x,y).feature_importances_)
```

```python
# Feature Importance using Filter Method (Chi-Square)
from sklearn.feature_selection import SelectKBest, chi2
X = df1.loc[:,df1.columns!='RainTomorrow']
y = df1[['RainTomorrow']]
selector = SelectKBest(chi2, k=10)
selector.fit(X, y)
X_new = selector.transform(X)
print(X.columns[selector.get_support(indices=True)])
```
